# RNA-Seq Annotation and Analysis
This code was used to annotate DESeq2 results for dnBMPR1A-FLAG vs Control cranial neural crest cells isolated by FACS at 8 somites stage (HH9+). 

Sample Preparation:

* RNA was extracted and cDNA libraries were prepared using the SMART-Seq v4 Ultra Low Input kit. 50 bp single-end reads were collected on an Illumina HiSeq sequencer in the Caltech Millard and Muriel Jacobs Genetics and Genomics Laboratory.


Required inputs for this script:

1. galGal6 genome annotation file (downloaded from Ensembl BioMart: https://m.ensembl.org/biomart/martview/31b9dc3565fd6ef5dd2af0381eed2461)
2. DESeq2 results comparing dnBMPR1A samples with Control Samples, values used for Volcano and MA plots
3. featureCounts table containing counts for each transcript in each sample (same input as into DESeq2 analysis), values used for Heatmap analysis

Script prepared by Mike Piacentino, January 2021

```python
# Import packages
import numpy as np
import pandas as pd
import bokeh
import bokeh.io
import bokeh.plotting
import holoviews as hv
import seaborn as sns
bokeh.io.output_notebook()
hv.extension('bokeh')
```

# Input and Compile Data

### Read in Genome Annotation File

Make sure that this genome annotation file contains the Ensembl Gene ID number, the gene name and description, and the GO term and GOSlim data

```python
# Call genome annotation file
galGal6 = pd.read_csv('../Inputs/galGal6_annotations.csv')
galGal6 = galGal6.rename(columns={"Gene stable ID": "GeneStableID", "Gene name": "GeneName"})  # Rename some columns for easier cleaning later

# Fix incomplete annotations in galGal6 genome
galGal6.loc[(galGal6['GeneStableID'] == 'ENSGALG00000030902'),'GeneName']='SNAI2'     # SNAI2 record on Ensembl does not list 'SNAI2' in 'Gene name'
galGal6.loc[(galGal6['GeneStableID'] == 'ENSGALG00000043204'),'GeneName']='PAX7'     # Pax7 record on Ensembl does not list 'PAX7' in 'Gene name'

# Missing entries from BioMart export of galGal6:
galGal6.loc[len(galGal6.index)] = ['ENSGALG00000042686', 'FOXD3', 'forkhead box D3 [Source:NCBI gene;Acc:395794]', 'NaN', 'NaN', 'NaN', 'NaN', 'NaN'] 

# Display dataframe
galGal6.head(2)
```

### Read in DESeq2 results, tidy data and filter based on pre-defined thresholds

* Mean expression ('baseMean') must be >= 1000
* False Discovery Rate (FDR, 'padj') must be < 0.05
* Fold change must be greater than 1.8 fold (equivalent to log2FoldChange > 0.85)

```python
# Import data
raw = pd.read_csv('../Inputs/dnBMPR1A_vs_Ctrl_DESeq2_results.csv')
full_df = pd.merge(raw, galGal6, left_on='Gene id', right_on='GeneStableID')
full_df['neglogpadj'] = -(np.log10(full_df['padj']))                            # Calculate -log(padj) for later plotting
full_df['FoldChange'] = 2**(full_df['log2FoldChange'])                          # Calculate fold change for later plotting
full_df = full_df.drop(columns=['KEGG Pathway and Enzyme ID','GO term accession',             # Don't need GO terms attached to these data for now
                                'GO term name', 'GOSlim GOA Accession(s)', 'GOSlim GOA Description', 
                                'Gene id', 'Unnamed: 0']).drop_duplicates()           
full_df = full_df.filter(['GeneStableID','GeneName','Gene description',                       # Sort columns in dataframes 
                                'baseMean','log2FoldChange','FoldChange','pvalue','padj','neglogpadj',
                                'Ctrl_rep1','Ctrl_rep2','dnBMPR1A_rep1','dnBMPR1A_rep2','lfcSE','stat'], axis=1)

# Filter dataset
filtered_df = (full_df[(full_df['baseMean'] >= 500)])                       # Filter on mean expression levels
filtered_df = (filtered_df[(filtered_df['padj'] < 0.05)])                    # Filter on FDR
dnbmpr1a_degs = (filtered_df[abs(filtered_df['log2FoldChange']) > 0.85])     # Note: 0.85 log2FoldChange is equivalent to 1.8 fold change, 1.0 log2FC = 2 fold
dnbmpr1a_up = (dnbmpr1a_degs[(dnbmpr1a_degs['log2FoldChange'] > 0.85)])      # Isolate upregulated genes
dnbmpr1a_down = (dnbmpr1a_degs[(dnbmpr1a_degs['log2FoldChange'] < -0.85)])   # Isolate downregulated genes

# Print out some results
print('dnBMPR1A Results:')
print(str(len(filtered_df))+' genes with padj < 0.05.')
print(str(len(dnbmpr1a_degs))+' DEGs with > 1.8 fold change.')
print(str(len(dnbmpr1a_up))+' upregulated DEGs.')
print(str(len(dnbmpr1a_down))+' downregulated DEGs.')
```

```python
# Generate list of Gene stable IDs that are missing a Gene name from the galGal6 annotations
missing_gene_names = dnbmpr1a_degs.loc[dnbmpr1a_degs['GeneName'].isna()]
missing_ids = missing_gene_names['GeneStableID'].tolist()
missing_ids
```

```python
# Make dictionary with Gene stable ID then gene name
unidentified_genes = {
    'ENSGALG00000028415': ['HES6-2'],
    'ENSGALG00000001136': ['HES5-3'],
    'ENSGALG00000001141': ['HES5-1'],
    'ENSGALG00000007710': ['RASL11AL'],
    'ENSGALG00000015684': ['DNAJC25'],
    'ENSGALG00000020788': ['CCDC141'],
    'ENSGALG00000003861': ['HERC4'],
    'ENSGALG00000031427': ['FMN2L'],
    'ENSGALG00000040546': ['ZNF217'],
    'ENSGALG00000003162': ['FAM134C'],
    'ENSGALG00000009453': ['MTARC1'],
    'ENSGALG00000020454': ['RIOX1']
                     }

# Go through and replace Gene name with names you've looked up
for i in unidentified_genes:
    for j in dnbmpr1a_degs['GeneStableID']:
        if i == j:
            dnbmpr1a_degs.loc[dnbmpr1a_degs.GeneStableID == j, 'GeneName'] = unidentified_genes[i][0]
    for k in full_df['GeneStableID']:
        if i == k:
            full_df.loc[full_df.GeneStableID == k, 'GeneName'] = unidentified_genes[i][0]
full_df = full_df

dnbmpr1a_up = (dnbmpr1a_degs[(dnbmpr1a_degs['log2FoldChange'] > 0.85)])      # Isolate upregulated genes
dnbmpr1a_down = (dnbmpr1a_degs[(dnbmpr1a_degs['log2FoldChange'] < -0.85)])   # Isolate downregulated genes

    
# Save out annotated DESeq Results
full_df.to_csv('dnBMPR1a_vs_Control_DESeq2_Full.csv')             # Save out full annotated DESeq results
dnbmpr1a_degs.to_csv('dnBMPR1a_vs_Control_DESeq2_DEGs.csv')       # Save out annotated Differentially Expressed Genes (DEGs)
```

### Read in RNASeq featureCounts and merge with genome anotations

```python
# Read in RNASeq feature counts from all samples to examine
counts_raw = pd.read_csv('../Inputs/20200828_dnBMPR1A_vs_Ctrl_featureCounts.txt', sep='\t')

# Merge genome annotations with feature couts dataframe
merged_counts = pd.merge(counts_raw, galGal6, left_on='Gene_ID', right_on='GeneStableID')
merged_counts = merged_counts.drop(columns=['KEGG Pathway and Enzyme ID','GO term accession', 
                                'GO term name', 'GOSlim GOA Accession(s)', 'GOSlim GOA Description',
                                'Gene_ID']).drop_duplicates()

# Go through and replace Gene name with names you've looked up
for i in unidentified_genes:
    for j in merged_counts['GeneStableID']:
        if i == j:
            merged_counts.loc[merged_counts.GeneStableID == j, 'GeneName'] = unidentified_genes[i][0]

# Display dataframe
merged_counts.head()
```

# Plotting and Data Visualization

## Heatmap analysis to cluster gene expression patterns

```python
# Filter out only DEGs for heatmap analysis
degs_list = dnbmpr1a_degs['GeneStableID'].tolist()
degs_counts = merged_counts.loc[merged_counts['GeneStableID'].isin(degs_list)]
conditions_list = ['GeneName',
            'Ctrl_rep1',                   # Populate this list with the column names you want to include in the heatmap
            'Ctrl_rep2',
            'dnBMPR1A_rep1',
            'dnBMPR1A_rep2',
             ]

# Assemble heatmap dataset for Seaborn formatting
heatmap_df = degs_counts.filter(conditions_list, axis=1).drop_duplicates()
melted = pd.melt(heatmap_df, id_vars=['GeneName'])
heatmap_data = pd.pivot_table(melted, values='value', 
                              index=['GeneName'], 
                              columns='variable')


# OPTIONAL: transpose data to rotate heatmap
heatmap_data = heatmap_data.transpose()

# Assemble heatmap
sns.set(font_scale=1.0)
g = sns.clustermap(heatmap_data
            # Clustering parameters:
                   ,method='complete', metric='euclidean', z_score=1
                   ,col_cluster=True, row_cluster=True
                   ,dendrogram_ratio=0.1
            # Plot customizations
                   ,cmap='RdBu_r'
#                    ,figsize=(6,20)
                   ,figsize=(22,6)
                   ,vmin=-1.75, vmax=1.75            
#                    ,cbar_pos=None
                   ,cbar_pos=(0.02, 0.93, 0.02, 0.1)
                   ,tree_kws=dict(linewidths=1.)
                   ,linewidths=0.005,linecolor='gray'
                  )

# OPTIONAL: save out heatmap
g.savefig('../Plots/dnBMPR1A vs Ctrl sigDEGS heatmap.png')
print(str(len(dnbmpr1a_degs))
      +' genes differentially expressed between dnBMPR1A and Ctrl.')
```

## Interactive bar plot for log2 fold change

```python
# Sort and separate DEGS df
sorted_degs = dnbmpr1a_degs.sort_values(['log2FoldChange'], ascending=True)
l2fc_threshold = 1
up, down = (sorted_degs.loc[sorted_degs['log2FoldChange'] >= l2fc_threshold],
        sorted_degs.loc[sorted_degs['log2FoldChange'] <= -l2fc_threshold])

# Assemble individual plots (up/down)
bar_up = hv.Bars(data=up, kdims=['GeneName'], vdims=['log2FoldChange']
                ).opts(tools=['hover'],color='#D22B25')
bar_down = hv.Bars(data=down, kdims=['GeneName'], vdims=['log2FoldChange']
                  ).opts(tools=['hover'], color='#6298C6')
bar_degs = bar_down*bar_up
# bar_degs = bar_down

# Format plot details
bar_degs = bar_degs.redim.label(GeneName= 'Gene Name', log2FoldChange='log\u2082(Fold Change)')
bar_degs.opts(height=400, width=900
              ,xrotation=90
              ,show_grid=True
             )
bar_degs.opts(fontsize={'title': 16, 'labels': 14, 'xticks': 12, 'yticks': 12})

# Display plot
print(str(len(bar_up)+len(bar_down))
      +' genes differentially expressed between dnBMPR1A and Ctrl with l2FC > '+str(l2fc_threshold))
bar_degs
```
